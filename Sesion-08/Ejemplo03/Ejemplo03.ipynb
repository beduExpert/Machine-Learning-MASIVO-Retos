En esta serie de ejemplos veremos descriptores que nos ayudarán a obtener información numérica de la imagen que estamos analizando. Un descriptor nos puede ayudar a hacer la tarea de clasificación mucho mas sencilla.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

imageOpenCV = cv2.imread("naranja.jpg", cv2.IMREAD_COLOR)
imagen = cv2.cvtColor(imageOpenCV, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(7,7))
plt.imshow(imagen)
```

## Histograma de colores.

Un histograma de color te permite conocer cuantos colores hay en una imagen y en que cantidad. Si por ejemplo, hay una dominancia de cierto color rojo y estas clasificando frutas, puedes saber que esa fruta es probablemente una manzana.

```python
def crearHistograma(imagen):
    histograma = np.zeros((256))
    imgEnArray = np.ravel(imagen)
    for counter in range(0, len(imgEnArray)):
        histograma[int(imgEnArray[counter])]+=1
        
    return histograma

histograma = crearHistograma(imagen[:,:,0])
print("Color mas usado:"+str(np.argmax(histograma))+". Usado "+str(np.max(histograma))+" veces.")
```

```python
histogramaRojo = crearHistograma(imagen[:,:,0])
histogramaVerde = crearHistograma(imagen[:,:,1])
histogramaAzul = crearHistograma(imagen[:,:,2])
```

```python
plt.bar(np.arange(len(histogramaRojo)),histogramaRojo, color='red')
plt.show()
plt.bar(np.arange(len(histogramaVerde)),histogramaVerde, color='green')
plt.show()
plt.bar(np.arange(len(histogramaAzul)),histogramaAzul, color='blue')
plt.show()
```

```python
imageOpenCV = cv2.imread("plastico.jpg", cv2.IMREAD_COLOR)
imagenPlastico = cv2.cvtColor(imageOpenCV, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(7,7))
plt.imshow(imagenPlastico)
```
